```python
import pandas as pd
from selenium import webdriver
from time import sleep
from random import randint
from selenium.webdriver.common.by import By
```

#### Khởi tạo driver

```python
driver = webdriver.Chrome()
driver.maximize_window()
```

#### Truy cập trang https://www.topcv.vn/tim-viec-lam-data-analyst

```python
driver.get("https://www.topcv.vn/tim-viec-lam-data-analyst")
sleep(randint(5,10))
```

#### Lấy thông tin từng job

```python
links = [] # Link job
title = [] # Tiêu đề
salary = [] # Mức lương
address = [] # Địa chỉ
company = [] # Tên công ty
```

```python
while True:
    try:
        # Lấy tất cả job ở từng trang
        jobs = driver.find_elements(By.CSS_SELECTOR, ".job-item-default.job-ta")
        sleep(randint(5,10))
        
        for job in jobs:
            # Lấy link từng job
            links.append(job.find_element(By.CSS_SELECTOR,"a").get_attribute("href"))
            # Lấy title từng job
            title.append(job.find_elements(By.CSS_SELECTOR, "a")[1].text)
            # Lấy mức lương từng job
            salary.append(job.find_element(By.CSS_SELECTOR, ".title-salary").text)
        # Kiểm tra xem có sang được trang sau không
        press = driver.find_element(By.CSS_SELECTOR, ".pagination li:nth-of-type(3)")
        # Nếu attribute class='' tức là còn sang được trang mới, class='disabled' tức là không còn trang mới
        if press.get_attribute('class') == '':
            press.click()
        else: 
            break
        sleep(randint(5,10))
    except:
        break
```

#### Truy cập vào từng link lấy thông tin cụ thể

```python
position = [] # Vị trí
description = [] # Mô tả công việc
```

```python
driver.get(links[0])
```

```python
driver.find_element(By.CSS_SELECTOR, ".premium-job-description__box--content").text
```

```python
for link in links:
    # Truy cập vào từng link
    driver.get(link)
    sleep(randint(5,10))
    
    # Lấy vị trị
    try:
        position.append(driver.find_element(By.CSS_SELECTOR,".box-general-group-info-value").text)
    except:
        try:
            position.append(driver.find_element(By.CSS_SELECTOR, ".general-information-data__value").text)
        except:
            position.append(None)
    
    # Lấy jd
    try:
        description.append(driver.find_element(By.CSS_SELECTOR,".job-description__item--content").text)
    except:
        try:
            description.append(driver.find_element(By.CSS_SELECTOR,".premium-job-description__box--content").text)
        except:
            description.append(None)
```

#### Lưu data xuống file

```python
data = pd.DataFrame(data=list(zip(title, position, description, salary, links)), 
                    columns = ['Tiêu đề', 'Vị trí', 'Mô tả', 'Mức lương', 'Link'])
```
